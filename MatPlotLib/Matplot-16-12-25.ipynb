```python
# working with sigmoid function

# Formula  for sigmoid function [ f(x) = 1/(1+e^(-x)) ]
# wrinting the sigmoid function using numpy ---- > np(x) == 1 / (1 + np.exp(-x))



def sigmoid(x):
    return 1/(1+np.exp(-x)) 
b = np.arange(10)

sigmoid(b)
 
```

```python
c = np.arange(100)
sigmoid(c)
```

```python
# Mean Squared Error [ average value , when do a square of that  -MSE ]
# It shows the difference between actual value and predicted value 
# Car 1ltr / 10 km  [ 1ltr / 7 km ]

# 10 - 7 = 3  # now square the value 3 

actual_values = np.random.randint(1, 10, 10)
predicted_values = np.random.randint(1, 10, 10)
print(actual_values)
print(predicted_values)
```

```python
# Creating a function for MSE
def mse(actual_values, predicted_values):
    return np.mean((actual_values - predicted_values)**2)

mse(actual_values, predicted_values)
```
