```python
# importing installed libraries 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Now we need dataset ( csv), for we can use built-in datasets from seaborn library 
len(sns.get_dataset_names()) # 12 datasets are available
sns.get_dataset_names()
```

```python
# loading titanic dataset from sns 
df = sns.load_dataset('titanic')
```

```python
df.info()
```

# In general we use some of the following charts / plots / graphs 

```python
# Types of charts : 
1. Line Plot : To show trend in data
2. Bar Plot : To show comparison between different categories
3. Pie Plot : To show comparison between different categories
4. Histogram : To show distribution of data
5. Scatter Plot : To show relationship between two variables
6. Box Plot : To show distribution of data and outliers
7. Violin Plot : To show distribution of data and outliers
8. Heatmap : To show correlation between variables
```

```python
df.columns
```

```python
#Working with line plot : i) Purpose of the plot , ii) required variables / parameters for each type of plot
#Purpose of the plot : To show trend in data
#Required variables / parameters for each type of plot : x-axis, y-axis, data
# x-axis : Age
# y-axis : Fare
# data : df
# To show trend in data : Line plot
# x-axis : Age
# y-axis : Fare
# data : df
# To show trend in data : Line plot
sns.lineplot(x='age',y='fare',data=df)
```

```python
# creating a line plot with matplotlib with same column, x = age, y = fare 
plt.plot(df['age'], df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# Working with bar plot : i) Purpose of the plot , ii) required variables / parameters for each type of plot
#Purpose of the plot : To show comparison between different categories
#Required variables / parameters for each type of plot : x-axis, y-axis, data
# x-axis : Age
# y-axis : Fare
# data : df
# To show comparison between different categories : Bar plot
# x-axis : Age
# y-axis : Fare
# data : df
sns.barplot(x='age', y='fare', data=df)
```

```python
# creating a bar plot with matplotlib with same column, x = age, y = fare
plt.bar(df['age'], df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# trying with barh plot
plt.barh(df['age'], df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# trying with histrogram plot
plt.hist(df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# trying histogram plot with bins
plt.hist(df['fare'], bins=10)
plt.show() # This should always be the last line of the code
```

```python
# using seaborn to create a histogram plot
sns.histplot(df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# Working with Scatter plot , with purpose , parameters
#Purpose of the plot : To show relationship between two variables
#Required variables / parameters for each type of plot : x-axis, y-axis, data
# x-axis : Age
# y-axis : Fare
# data : df
# To show relationship between two variables : Scatter plot
# x-axis : Age
# y-axis : Fare
# data : df
sns.scatterplot(x='age', y='fare', data=df)

```

```python
# trying scatterplot with matplotlib 
plt.scatter(df['age'], df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# working with box plot 
sns.boxplot(df['fare'])
plt.show() # This should always be the last line of the code
```

```python
# working with violin plot
sns.violinplot(df['fare'])
```

```python
# working with heatmap
sns.heatmap(df.corr())

# This should always be the last line of the code
```
