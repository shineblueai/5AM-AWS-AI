```python
# All of tensor operations 

one_dim_tens = tf.constant([1,2,3,4,5])
second_one_dim_tens = tf.constant([6, 7, 8, 9, 10])
```

```python
one_dim_array = np.array([1, 2, 3, 4, 5])
second_dim_array = np.array([6, 7, 8, 9, 10])
result_array_add = np.add(one_dim_array, second_dim_array)
result_array_sub = np.subtract(one_dim_array, second_dim_array)
result_array_mul = np.multiply(one_dim_array, second_dim_array)
result_array_div = np.divide(one_dim_array, second_dim_array)
print(result_array_add, result_array_sub, result_array_mul, result_array_div)
```

```python
result_add = tf.add(one_dim_tens, second_one_dim_tens)
result_mul = tf.multiply(one_dim_tens, second_one_dim_tens)
result_sub = tf.subtract(one_dim_tens, second_one_dim_tens)
result_div = tf.divide(one_dim_tens, second_one_dim_tens)

print(result_add)
print(result_mul)
print(result_sub)
print(result_div)

```

```python
nn = tf.convert_to_tensor(one_dim_array, dtype=tf.float32)
type(nn)
```

```python
# performing a matrix mul in numpy 
s_mat_1 = np.arange(12).reshape(3, 4)
s_mat_2 = np.arange(12,24).reshape(4, 3)
mat_mul_np = np.matmul(s_mat_1, s_mat_2)
print(mat_mul_np)
```

```python
s_mat_1 @ s_mat_2
```

```python
# performing a matrix multiplication using tensorflow 

# converting an array into tensor 
t_mat_1 = tf.convert_to_tensor(s_mat_1, dtype=tf.float32)
t_mat_2 = tf.convert_to_tensor(s_mat_2, dtype=tf.float32)
t_mat_1 @ t_mat_2
```

```python
# Performing a dot product in numpy 
array_dot_p = np.dot(one_dim_array, second_dim_array)
print(array_dot_p)
```

```python
# performing dot product using tensorfow 
tens_dot_p = tf.tensordot(one_dim_tens, second_one_dim_tens, axes=1)
print(tens_dot_p)
```

```python
# Creating a variable using pytorch  , tensor 
mm = torch.tensor([1, 2, 3, 4, 5])
xx = torch.tensor([6, 7, 8, 9, 10])
```

```python
# now performing math ops using pytorch 

result_torch_add = torch.add(mm, xx)
result_torch_sub = torch.sub(mm, xx)
result_torch_mul = torch.mul(mm, xx)
result_torch_div = torch.div(mm, xx)
print(result_torch_add, result_torch_sub, result_torch_mul, result_torch_div)
```

```python
# performing dot product using pytorch 

print(np.dot(s_mat_1, s_mat_2))
print(tf.tensordot(one_dim_tens, second_one_dim_tens, axes=1))
print(torch.dot(mm, xx))
```
