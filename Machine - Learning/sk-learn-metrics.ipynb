```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
#. Mean Absolute Error ( MAE)
# Definition :It is the average of the absolute differemces betwen predicted and actual values 
# Purpose / where do use : Linear Regression, Decision Tree, Random Forest Regressor
# Import statement : from sklearn.metrics  import mean_absolute_error

from sklearn.metrics import mean_absolute_error

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
mae = mean_absolute_error(y_true, y_pred)
print("Mean Absolute Error:", mae)
```

```python
# Mean Squared Error (MSE)
# Definition : It is the average of the squared differences between predicted and actual values.
# Purpose / where do use : Linear Regression, Decision Tree, Random Forest Regressor
# Import statement : from sklearn.metrics  import mean_squared_error

from sklearn.metrics import mean_squared_error
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Root Mean Squared Error (RMSE)
# Definition : It is the square root of the average of the squared differences between predicted and actual values.
# Purpose / where do use : Linear Regression, Decision Tree, Random Forest Regressor
# Import statement : from sklearn.metrics  import mean_squared_error

from sklearn.metrics import mean_squared_error
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print("Root Mean Squared Error:", rmse)
```

```python
# r2_score [ It measures how well the model explains the variance in the target variable ]
# Definition : It is the coefficient of determination, which measures the proportion of the variance in the dependent variable that is predictable from the independent variable.
# Purpose / where do use : Linear Regression, Decision Tree, Random Forest Regressor
# Import statement : from sklearn.metrics  import r2_score
# when we make prediction at that time instead of prediction , we will use the word score(real data)

from sklearn.metrics import r2_score
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])
r2 = r2_score(y_true, y_pred)
print("R-squared:", r2)
```

```python
# MAE, MSE, RMSE, R2_score, 
# Classification Report : It is used to evaluate the performance of a classification model.
# Import statement : from sklearn.metrics  import classification_report
from sklearn.metrics import classification_report
y_true = np.array([1, 0, 1, 1, 0, 1])
y_pred = np.array([0, 0, 1, 1, 0, 1])
report = classification_report(y_true, y_pred)
print("Classification Report:\n", report)
```
