```python
import pandas as pd
import numpy as np

my_list = [1, 2, 3, 4, 5]
#.        

# finding mean value for above list
mean_value = np.mean(my_list)
print("Mean value:", mean_value)
```

```python
# min value in above list is : 1 
# max value in above list is : 5
min_value = np.min(my_list)
max_value = np.max(my_list)
print("Min value:", min_value)
print("Max value:", max_value)
```

```python
# now if I sub one by one value from mean to find the answer
# 1 - 3.0 = -2.0
# 2 - 3.0 = -1.0
# 3 - 3.0 = 0.0
# 4 - 3.0 = 1.0
# 5 - 3.0 = 2.0
# so the answer is : [-2.0, -1.0, 0.0, 1.0, 2.0]
my_list = [1, 2, 3, 4, 5]
mean_value = np.mean(my_list)
print("Mean value:", mean_value)
```
