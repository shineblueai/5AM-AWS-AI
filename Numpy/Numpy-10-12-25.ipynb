# What is the Data Strcture of Numpy 
A. The datastructure of numpy as known as nd array , which means , n-dimesnional array 

```python
# Crating an array 

a = np.array(5)
print(a)
print(type(a))
```

# A single number  is known  as a  scalar [ 5 ] -> has no dimension [0-zero]
# A group of scalers is known as a vector [5,4,3,2] - has 1 dimension [ 1- dim]
# A Group of vectors is known as a matrix [ [5,4,3,2], [10,9,9,7]] -> has a [2 - dim]
# A group of matrices is known as a tensor -has a min of [3-DIM]

```python
# Findig more information about a numpy array [a- attributes and methods]
print("datatype of array a:", a.dtype)
print("number of dimensions of array a:", a.ndim)
print("shape of array a:", a.shape) # {number of rows, number of columns}
print("size (total number of elements) of array a:", a.size)
print("itemsize (size in bytes of each element) of array a:", a.itemsize)
print("nbytes (total size in bytes) of array a:", a.nbytes)
```

```python
b =np.array([1,2,3,4,5])
print("datatype of array b:", b.dtype)
print("number of dimensions of array b:", b.ndim)
print("shape of array b:", b.shape) # {number of rows, number of columns}
print("size (total number of elements) of array b:", b.size)
print("itemsize (size in bytes of each element) of array b:", b.itemsize)
print("nbytes (total size in bytes) of array b:", b.nbytes)
```

```python
# Creating a 2D array

c = np.array([ [1,2,3], [4,5,6] ])
print("datatype of array c:", c.dtype)
print("number of dimensions of array c:", c.ndim)
print("shape of array c:", c.shape) # {number of rows, number of columns}
print("size (total number of elements) of array c:", c.size)
print("itemsize (size in bytes of each element) of array c:", c.itemsize)
print("nbytes (total size in bytes) of array c:", c.nbytes)
```

```python
# Crating a 3d array
d = np.array([ [[[ [1,2,3], [4,5,6] ] ] ]])
print("datatype of array d:", d.dtype)
print("number of dimensions of array d:", d.ndim)
print("shape of array d:", d.shape) # {number of rows, number of columns}
print("size (total number of elements) of array d:", d.size)
print("itemsize (size in bytes of each element) of array d:", d.itemsize)
print("nbytes (total size in bytes) of array d:", d.nbytes) 
```
