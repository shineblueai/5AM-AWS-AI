```python
# Reshape command in numpy
array = np.arange(1,13) # n-1
print("Original Array is :\n", array)
print(array.ndim)
print(array.shape)
```

```python
array.reshape(3,4)
print("Reshaped Array is :\n", array.reshape(3,4))
print(array.ndim)
print(array.shape)
```

```python
array.reshape(4,3)
print("Reshaped Array is :\n", array.reshape(4,3))
print(array.ndim)
print(array.shape)
```

```python
array.reshape(2,2,3)
print("Reshaped Array is :\n", array.reshape(2,2,3))
print(array.ndim)
print(array.shape)
```
