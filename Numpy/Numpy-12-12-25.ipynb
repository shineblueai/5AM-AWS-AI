```python
# Vector Operations 
first_vector = np.array([1, 2, 3, 4, 5])
second_vector = np.array([6, 7, 7, 9, 10])
print(first_vector + second_vector)
print(first_vector - second_vector)
print(first_vector * second_vector)
print(first_vector / second_vector)
```

```python
#  array multiplication
first_vector = np.array([1, 2, 3, 4, 5])
second_vector = np.array([6, 7, 8, 9, 10])
result_vector = np.multiply(first_vector, second_vector)
print(result_vector)
```

```python
# Matrix Operations
first_matrix = np.array([[1, 2, 3], [4, 5, 6]])
second_matrix = np.array([[7, 8, 9], [10, 11, 12]])
print(first_matrix + second_matrix)
print(first_matrix - second_matrix)
print(first_matrix / second_matrix)
```

```python
# Matrix Multiplication

# Number of column in first matrix = Number of rows in second matrix
first_matrix = np.array([1, 2, 3,]) # (1,3)
second_matrix = np.array([12,13,14]) # (1,3 )


array_multiply = np.multiply(first_matrix, second_matrix)   
print(array_multiply)
# Matrix Multiplication
matrix_multiply = np.matmul(first_matrix, second_matrix)
print(matrix_multiply)
```

```python
new_array = np.array([[1,2,3], [4,5,6]])
print(new_array)
print(new_array.shape)
print()
now_array = np.array([[10,20,30], [40,50,60]])
print(now_array)
print(now_array.shape)
```

```python
# performing matrix multiplication (m x n)
matrix_multiply = np.matmul(new_array, now_array)
print(matrix_multiply)
```

```python
# transpose of matrix
tranposed_array = np.transpose(now_array)
print(tranposed_array)
print(tranposed_array.shape)
```

```python
# performing a matrix multiplication
matrix_multiply = np.matmul(new_array, tranposed_array)
print(matrix_multiply)
```
