```python
# Working with Python List DataType

#R1 : TO use Python List we need to use [ ]
#R2: EVerything insided a list is known as an item
#R3, Every item has to be sep by a comma
#4, If we are using string datatype we need to enclose in a pair of '' or ""
#5. List is mutable datatype
```

```python
# Creating a list

my_first_numbers = [1, 2, 3, 4, 5]
my_second_numbers = [ 6, 7, 8, 9, 10]
print(my_first_numbers, my_second_numbers)
print() # Whenever we type an empty print statement we will have a blank line
print(type(my_first_numbers))
print(type(my_second_numbers))
```

```python
# Working with some of the list methods. Here are some of them :

# Append Method , [ .append method, adds new items to the very end of the list]
# Syn:variable_name.append(item)
my_first_numbers.append(11)
print(my_first_numbers)

```

```python
# Working with second example
my_first_numbers.append(100)
print(my_first_numbers)
```

```python
# using len() function : It provides length of items [ number of items]
print(len(my_first_numbers))

# Syn : len(variable_name)
len(my_first_numbers)
```
