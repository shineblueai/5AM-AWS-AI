```python
# if -elif - else example

A = 100
B = 100

if A > B:
    print("A is greater than B")
elif A < B:
    print("A is less than B")
else:
    print("A is equal to B")
```

```python
# Working with For Loops,
# writing 2 table using for loop
# range function : Generates a sequence of numbers [ start, stop)]
for i in range(1, 11):
    print(f"2 x {i} = {2 * i}")     # Formatted string literals (f-strings)
```

```python
# working with range function

# Range functuion purpose : to generate a sequence of numbers [ start, stop)

# printing 1 to 100 numbers using range function

for c  in range(1, 5):
    print(c)
```

```python
# Working with While Loops : 

a = 1
b = 5

while a <= b:
    print(a)
    a += 1
```
