```python
# VAE Sampling Demo (conceptual)
import numpy as np

def reparameterize(mean, log_var):
    epsilon = np.random.normal(size=mean.shape)
    return mean + np.exp(0.5 * log_var) * epsilon

# Example latent space sampling
mu = np.array([0.0, 0.0])
logvar = np.array([0.1, 0.1])
sample = reparameterize(mu, logvar)
print("Sampled latent vector:", sample)
```
