```python
# Simple Rule-Based Chatbot (expandable)
import re

def respond(user_input):
    user_input = user_input.lower().strip()
    if re.search(r"\b(hi|hello|hey)\b", user_input):
        return "Hello! How can I help you with EIT Academy courses?"
    elif "course" in user_input and "ai" in user_input:
        return "We offer AI+ML+GenAI course (45+ hours) by Phani Rajendra."
    elif "trainer" in user_input:
        return "Our lead trainer is Phani Rajendra with 25+ years of experience."
    elif "contact" in user_input or "website" in user_input:
        return "Visit us at https://eitacademy.in or call 9170222195."
    else:
        return "I'm still learning! Can you rephrase or ask about courses, trainers, or contact info?"

# Demo loop
print("ðŸ¤– EIT Bot: Hi! Ask me about courses.")
while True:
    user = input("You: ")
    if user.lower() in ['exit', 'quit', 'bye']:
        print("ðŸ¤– Goodbye!")
        break
    print("ðŸ¤–", respond(user))
```
