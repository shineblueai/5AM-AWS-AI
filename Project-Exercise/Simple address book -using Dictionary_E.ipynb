```python
# Simple Address Book
address_book = {}

def add_contact(name, phone, email):
    address_book[name] = {'phone': phone, 'email': email}

def get_contact(name):
    return address_book.get(name, "Contact not found")

# Add entries
add_contact("Alice", "123-4567", "alice@example.com")
add_contact("Bob", "987-6543", "bob@example.com")

# Retrieve
print(get_contact("Alice"))
print(get_contact("Charlie"))  # Not found
```
