```python
# Few-Shot Prompting Simulation
examples = [
    ("What is the capital of France?", "Paris"),
    ("What is 5 + 7?", "12"),
    ("Who wrote 'Romeo and Juliet'?", "William Shakespeare")
]

def few_shot_prompt(query):
    prompt = "\n".join([f"Q: {q}\nA: {a}" for q, a in examples])
    prompt += f"\nQ: {query}\nA:"
    # In real LLM, send `prompt` to model
    return f"(Simulated) {prompt}"

print(few_shot_prompt("What is the boiling point of water in Â°C?"))
```
